```python
import pandas as pd
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt

# 샘플 데이터 생성
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, np.nan, 30, 22, np.nan],
    "Score": [85, 90, np.nan, 88, 95],
    "City": ["Seoul", "Busan", np.nan, "Incheon", "Daegu"]
}

df = pd.DataFrame(data)

print("원본 데이터")
print(df)

# 1. 막대 그래프(bar)로 결측치 확인
msno.bar(df)
plt.show()

# 2. 매트릭스(matrix) 형태로 결측치 확인
msno.matrix(df)
plt.show()

# 3. 히트맵(heatmap)으로 결측치 상관관계 확인
msno.heatmap(df)
plt.show()
```
